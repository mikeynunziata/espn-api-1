```python
# Football API
from espn_api.football import League

#import necessary packages
import pandas as pd
import numpy as np
import requests
```

```python
# Provide the league ID and year you want to view the data from
league_id = 214643
year = 2010

# Create the url using the league ID & year
url = "https://fantasy.espn.com/apis/v3/games/ffl/leagueHistory/" + \
      str(league_id) + "?seasonId=" + str(year)

# using requests to get url with proper data (depends on the data you want to view -> "mTeam" = League Standings)
r = requests.get(url, params={"view": "mTeam"})

# Creating the json file so that we can begin to pull data
d = r.json()[0]
```

```python
df = [[
        team['primaryOwner'], team['location'] + ' ' + team['nickname'], team['record']['overall']['wins'],
        team['record']['overall']['losses'], team['record']['overall']['ties'], team['record']['overall']['percentage'],
        team['record']['overall']['pointsFor'], team['record']['overall']['pointsAgainst'], 
        team['transactionCounter']['acquisitions'], team['transactionCounter']['trades'], 
        team['transactionCounter']['drops']
    ] for team in d['teams'] ]
df = pd.DataFrame(df, columns=['OID', 'Team Name', 'W', 'L', 'T', 'PCT', 'PF', 'PA', 'Moves', 'Trades', 'Drops'])
df.insert(2, "Year", d['seasonId'])
df.insert(10, "PF/G", df['PF'] / (df['W'] + df['L'] + df['T']))
df.insert(11, "PA/G", df['PA'] / (df['W'] + df['L'] + df['T']))
df.insert(12, "DIFF", df['PF/G'] - df['PA/G'])

df
```
